```python
# histograms
# data
penguins = sns.load_dataset("penguins")

sns.histplot(data=penguins, x="bill_length_mm", hue="island", multiple="stack")
```

```python
sns.kdeplot(data=penguins, x="bill_length_mm", hue="island", multiple="stack")
```

```python
sns.displot(data=penguins, x="bill_length_mm", hue="species", col="species")
```

```python
import matplotlib.pyplot as plt
```

```python
# seaborn have a great interaction with matplotlib and to know that API may help to create greater plots
f, axs = plt.subplots(1, 2, figsize=(8, 4), gridspec_kw=dict(width_ratios=[6, 3]))
sns.scatterplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species", ax=axs[0])
sns.histplot(data=penguins, x="species", hue="species", shrink=.8, alpha=.8, legend=False, ax=axs[1])
f.tight_layout()
```
